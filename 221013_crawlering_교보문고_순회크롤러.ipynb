# 순회크롤러

- 같은 양식의 url를 순회하면서 자료를 수집해오는 crawler
- 원 url crawler를 제작한 다음, 완성된 crawler를 range에 넣어서 만듬
- range를 어디부터 돌릴지에 대한 파악이 무엇보다 중요함 

```python
1. n page 접근
2. source code 긁어오기
3. parsing
4. 원하는 data 추출
5. .txt에 저장
6. 1번으로 이동
# 금일(221013)기준 45번 range을 돌려서
# next page xpath를 마지막에 클릭하게 하시오
```

```python
driver = webdriver.Chrome('chromedriver')
time.sleep(3)

driver.get("https://product.kyobobook.co.kr/bestseller/online?period=001")
time.sleep(3) 
# 스크롤을 내리지 않으면 오류가 발생. selenium Scroll Down키를 사용하면 해결.
driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")

title_kyobo = list()
price_kyobo = list()
author_kyobo = list()

for i in range(45):
    #source code 가져오기
    source = driver.page_source
    # parsing
    parsed_source = BeautifulSoup(source, "html.parser")
    # data 추출

    span_title_list = parsed_source.find_all("span", class_="prod_name")
        # all_code에서 span tag이면서 classrk prod_name인 tag만 save
    for span_list in span_title_list:
        print(span_list.text)
        # title만 추출해서 print()
        title_kyobo.append(span_list.text)
   
    span_price_list = parsed_source.find_all("span", class_="price")
    for span_price in span_price_list:
        print(span_price.text)
        price_kyobo.append(span_price.text)
        
    span_author_list = parsed_source.find_all("span", class_="prod_author")
    for span_author in span_author_list:
        print(span_author.text)
        author_kyobo.append(span_author.text)
        
    driver.find_element(By.XPATH, '//*[@id="tabRoot"]/div[2]/div[1]/button[2]').click()
    time.sleep(3)

driver.close()
```

```python
# txt에 저장
# codecs를 이용하여 kyobo_result.txt를 만드시오
# csv로 바꿔처리
# 순위, 제목, 저자, 가격
# 순차적으로 부여

kyobo_result = codecs.open("C:/Users/Playdata/python_basic/crawler/kyobo_result.text", mode="w")
# %s.txt %(날짜)<-'관련 코드' 로 정하면 매일 그 날짜에 따른 텍스트를 생성
kyobo_result.write("순위,제목,저자,가격\n")

for i in range(len(title_kyobo)):
    kyobo_result.write("%s등,%s,%s, %s\n" % ((i+1), 
                               title_kyobo[i].replace(",",""), 
                               author_kyobo[i].replace(",",""),
                               price_kyobo[i].replace(",","")))

kyobo_result.close()
```
