### 산술연산

- dataframe 과 스칼라 값 산술연산
- dataframe 과 series 간의 산술연산
- dataframe 간의 산술연산
    - 컬럼, 로우 인덱스를 기준으로 연산 수행
    - 공통으로 존재하지 않는 경우 NaN 반환
    - fill_value 인자 값을 통해 NaN이 아닌 값으로 대체 가능
- 연산의 종류
    - 더하기 : +, add() 메서드
    - 빼기 : -, sub() 메서드
    - 곱하기 : *, mul() 메서드
    - 나머지만 반환 : %
    - 몫만 반환 : //

```python
# 컬럼명을 서울, 경기, 인천으로
# 로우명을 a,b,,c로 가지고
# 0부터 1씩 증가하는 정수 값을 데이터로 가지는 DataFrame
df1 = pd.DataFrame(np.arange(9).reshape(3, 3), 
                   index=list('abc'), 
                   columns = ['서울', '경기', '인천'])
df1
```

```python
# 컬럼명을 서울, 경기 ,인천 ,세종 강원으로
# 로우명을 a,b,c,d 로 가지고 0부터 1씩 증가하는 정수값을 데이터로 가지는 df
df2 = pd.DataFrame(np.arange(20).reshape(4,5),
                 index = list('abcd'),
                  columns = ['서울','경기','인천','세종','강원'])
df2
```

```python
# 데이터프레임간의 더하기 연산
# 결과 해석 : 공통 컬럼, 로우인 데이터만 정상 더하기, 아닌 부분은 NaN
df1 + df2
```

```python
# fill_value : add method parameter
df1.add(df2, fill_value=0)
```

```python
# dataaframe 간의 빼기 연산
df1 - df2
```

```python
df1.sub(df2,fill_value=0)
```

```python
# 데이터프레임간의 곱하기 연산
df1 * df2
```

```python
# .mul로 처리. 곱연산시 대상 값이 없는 부분을 1로 간주하고 계산
df1.mul(df2, fill_value=1)
```

```python
# 데이터프레임 나누기 연산
df1/df2
```

```python
# .divide를 이용해 값이 없는 부분을 1로 간주해 나누시오
df1.divide(df2, fill_value=1)
```

### DataFrame과 Series 간의 연산

- 기본적인 동작은 Series 객체의 인덱스를 DataFrame 객체의 컬럼 인덱스와 매핑하여 브로드캐스팅과 유사하게 연산 수행
- 두 객체 간의 공통된 인덱스가 아닌 대상은 NaN 값으로 대입
- 메서드를 사용하여 연산을 수행할 때는 axis 파라미터를 통해 연산을 적용할 축 지정(0:행, 1:열)
- 연산의 종류
    - 더하기 : +, add() 메서드
    - 빼기 : -,  sub() 메서드
    - 곱하기 : *, mul() 메서드

```python
# 컬렴명 : a,b,c,d
# 로우명 : 2010,2011,2012
# 데이터 : 0부터 1씩 증가하는 정수
df = pd.DataFrame(np.arange(12).reshape(3,4),
                 index=[2010,2011,2012],
                 columns = list('abcd'))
df
```

```python
# 첫 번째 행을 추출
# s1 : 라벨인덱스가 적용된 시리즈
s1 = df.iloc[0]
s1
```

```python
# Sreies(라벨인덱스 a,b,c,d)와 df의 결합기준 : 이름이 일치할 때 더하기
df + s1
```

```python
# 데이터프레임
# 컬럼명 : a,b,c,d,e
# 데이터 : 20개의 0
df2 = pd.DataFrame(np.zeros(20).reshape(4,5),
                columns=list('abcde'))
df2
```

```python
# 기본인덱스를 가진 시리즈
# 데이터 0,1,2,3,4
s2 = pd.Series(np.arange(5))
s2
```

```python
# 열 단위 산술 연산
# 기본 동작  : Series 로우 인덱스를 df의 컬렁이세 매핑
# 결과해석
df2.sub(s2)
```

```python
# aixs=0 : 컬럼과 시리즈의 라벨인덱스가 아닌 로우와 시리즈의 라벨인덱스 일치시 연산
# 결과해석 : idx 4번의 결과만 NaN(s2의 4번 인덱스)
df2.sub(s2, axis=0)
```

```python
# df의 컬럼에 없는 인덱스를 가진 Sreies
# 인덱스 : a,b,c
# 데이터 : 3,3,3
s3 = pd.Series([3,3,3],index=list('abc'))
s3
```

```python
# 공통되지 않은 인덱스값과 컬럼값 -> NaN
df - s3
```

```python
df.sub(s3) # Series 에서는 fill_value를 제외함
```
