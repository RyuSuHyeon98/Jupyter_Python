```python
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
import time 
from selenium.webdriver.common.by import By 
```

```python
import codecs
```

```python
driver = webdriver.Chrome('chromedriver')
time.sleep(3)

driver.get("https://www.aladin.co.kr/shop/common/wbest.aspx?BranchType=1&start=we")
time.sleep(3)
```

```python
source = driver.page_source

parsed_source = BeautifulSoup(source, "html.parser")
```

```python
len(parsed_source.find_all("li"))
```

```python
# div tag 이면서 class가 ss_book_box인 것
len(parsed_source.find_all("div",class_="ss_book_box"))
```

```python
# bs4.resultest 자료형은 find_all 을 받을 수 없음(list로 취급하기 때문)
# 따라서 내부 요소를 개별적으로 처리해야 함
# 1. indexing
# 2. range 내부에서 돌아가면서 대입,처리
div_book_list = parsed_source.find_all("div",class_="ss_book_box")
div_book_list
```

```python
book_li_list = div_book_list[1].find_all("li")
print(book_li_list[1].text)
```

```python
# 50개 div를 반복문으로 한권씩 return
for book in div_book_list:
    # li만 조회
    li_list = book.find_all("li")
    # 변수의 1,2,3 번째만 print()로 찍어서 제목, 저자, 가격 순으로 출력
    for i in range(1,4):
        print(li_list[i].text)
```
